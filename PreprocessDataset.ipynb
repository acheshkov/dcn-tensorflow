```python
import numpy as np
import tensorflow as tf
import utils
import dataset as ds
```

```python
max_sequence_length = 160

dataset = ds.getDataset(["./train_task_b.csv"], max_sequence_length)
iterator = dataset.make_one_shot_iterator()
next_element = iterator.get_next()

fout = open('train_dataset_160.csv', 'w')
with tf.Session() as sess:
    for step_ in range(70000):
        try:
            start_true, end_true, doc, que, doc_v, que_v = sess.run(next_element)
            doc_dec = " ".join(utils.b2s(doc))
            que_dec = " ".join(utils.b2s(que))
            line = ";".join([str(start_true),str(end_true), doc_dec, que_dec])
            fout.write(line + '\n')
            if step_ % 150 == 0: print(step_)
        except tf.errors.OutOfRangeError:
            print("End of dataset")  # ==> "End of dataset"
            break
        except: 
            print("unknown error")
print("Finish")
fout.close()
```
